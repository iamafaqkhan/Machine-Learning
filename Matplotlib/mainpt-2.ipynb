Categorical Plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
df['sex'].unique()
```

```python
sns.countplot(x=df['smoker'], hue=df['sex'])
```

Bar Plot

```python
sns.barplot(x=df['sex'], y=df['total_bill'] , palette='rainbow')
```

```python
import numpy as np
sns.barplot(x=df['sex'], y=df['tip'] , palette='rainbow', estimator=np.sum)
```

Box Plot

```python
sns.boxplot(x=df['day'], y=df['tip'], data=df, palette='rainbow')
```

Voilin Plot

```python
sns.violinplot(x=df['smoker'], y=df['day'], palette='rainbow')
```

Strip Plot --> Shows a scatterplot, one variable is categorical

```python
sns.stripplot(x='day', y='total_bill', data=df, palette='rainbow')
```

Swarm Plot

```python
sns.swarmplot(x='day', y='total_bill', data=df)
```

Combining Swarmplot and Violin Plot

```python
sns.violinplot(x='tip', y='day', data=df, palette='rainbow')
sns.swarmplot(x='tip', y='day', data=df,)
```
