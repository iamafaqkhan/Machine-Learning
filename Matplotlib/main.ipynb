```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Matplotlib Basics

```python
x = np.linspace(0,10,5)
x
y = x ** 2
y
```

```python
plt.plot(x, y)
plt.title("Hi")
plt.xlabel("xaxis")
plt.ylabel("ylabel")
```

Subplots

```python
plt.subplot(2,2,1)
plt.plot(x**2,y**3)
plt.subplot(2,2,2)
plt.plot(x**4,y**2)
plt.subplot(2,2,3)
plt.plot(x,y**3)
plt.subplot(2,2,4)
plt.plot(x**2,y)
```

Object Oreinted Way, A method in which we can maipulate a canvas

figsize --> A tuple of width and height of the figure in inches DPI is the Dot-per-Inche.

```python
fig = plt.figure(figsize=(6,10), dpi=75)
fig.add_axes([0.1,0.1,1,1])
plt.plot(x,y)
anoplott = fig.add_axes([0.5, 0.5,0.5,0.5])
anoplott.plot(x, y)
```

Seaborn Section

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataframe = sns.load_dataset('tips')
dataframe
```

```python
dataframe['size'].unique()
```

```python
plt.subplot(2,2,1)
sns.histplot(dataframe['tip'], color='green', bins=20, kde=True)
plt.subplot(2,2,2)
sns.histplot(dataframe['total_bill'], color='pink', bins=20, kde=True)
plt.subplot(2,2,3)
sns.histplot(dataframe['sex'], color='blue', bins=20, kde=True)
plt.subplot(2,2,4)
sns.histplot(dataframe['day'], color='red', bins=20, kde=True)
```

Joint Plot

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = dataframe, kind='scatter')
```

Pair Plot

```python
sns.pairplot(dataframe, hue='size', palette='rainbow')
```

Rugg Plot

```python
sns.rugplot(dataframe)
```
